Apache Beam: Googles pipeline logic defintion model (where you define various activities such as data ingestion,
             cleansing, transofrmation and persistence (ETL tasks)) supports python, jave, Go and other SDKs
DataFlow: Execution engine (executes pipelines) for Apache Beam
Typical ApacheBeam pipeline: Source-->PipelineI/O1-->Collection1-->Transform1-->Collection2-->Transform2-->PipelineIO2-->Sink

Pcollection: In memory working dataset in the pipeline (ex: Python df or Java collection) - when pipeline ix executed in a parallel processing framework (such as Spark) the execution engine manages distribution of data across processing nodes - holder of data between pipeline steps

Transform: Typically takes input from one or more collections and transforms them and send output into one or more collections downstrea - used for cleanseing data record by record - can be executed in parallel across multiple processing nodes - (joining, sum, avg, groupings, aggreations, cutom agregs etc.) - Ptyhon or Java

PipielineI/O: inteface between data sources and sinks - provides libraries and interfaces to interact with variousstandard external data sources and sinks - Supports interface for popular databases and message ques - supports flat files (such as FileIO, TextIO, XMLIO), supports message ques (such as Amazon Kinesis, Apache Kafka, Cloud Pub/Sub) and supports interfaces for poular DBs (such as Cassandra, HBase, Elasticsearch, Bigquery

Runner: used to execute an ApacheBeam pipeline - runner is an interface that is provided to Apche beam by many data processing platforms where pipelines can be executed - Apche translates pipeline to runner specific code and execute code on the runner (for ease of scalability and portability) - runner example (DirectRunner (local PC native runner), SparkRunner(executes pipline on Apache Spark), DataFlow runner (executes pipline on Cloud DataFlow) -  so you only write once but execute it on multiple platforms

```python
#before executing the below I installed "apache_beam" and "apache-beam[interactive]"
#import print library
import logging
#import apache beam library
import apache_beam as beam
#import pipeline options
from apache_beam.options.pipeline_options import PipelineOptions
#set log level to info
root = logging.getLogger()
root.setLevel(logging.INFO)
#create a pipeline
p = beam.Pipeline(options=PipelineOptions())
#Create a PCollections from an in-memory list
lines = (
        p
        | 'Create Sample' #name of this step in the pipeline
                >> beam.Create(['training exercise', 'simple exercise', 'good training'])) #>> denotes action of step
#output of above pipeline is stored in the "lines" variable

#count words using a pipeline and create a results PCollections
linecount = (
        lines #using lines from step before
        | 'Count Lines'
                >> beam.CombineGlobally(beam.combiners.CountCombineFn())) #count lines in lines Pcollection

#print
(linecount
 | 'printing to log'
        >> beam.ParDo(lambda c: logging.info('*** Total Lines: %s ***\n',c)))

#run the pipeline
result = p.run()

#wait until pipeline processing is complete - keeps program running until entire pipleine has compelted

result.wait_until_finish()
```
